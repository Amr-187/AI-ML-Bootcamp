**Train Test Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X6=df.drop(['name','selling_price'],axis=1)
y6=df['selling_price']
```

```python
X6_train, X6_test, y6_train, y6_test = train_test_split(X6,y6,test_size=0.2,random_state=42)
```

```python
model6=LinearRegression()
model6.fit(X6_train,y6_train)
```

```python
y6_pred=model6.predict(X6_test)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(y6_test,y6_pred,color='green',label='Predicted vs. Actual')
plt.plot([min(y6_test),max(y6_test),min(y6_test),max(y6_test)],color='blue',linestyle='-',label='Completely Accurate Prediction')
plt.xlabel('Real Price')
plt.ylabel('Predicted Price')
plt.title('Real vs. Predicted Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

mse=mean_squared_error (y6_test,y6_pred)
rmse = np.sqrt(mse)
mae= mean_absolute_error(y6_test,y6_pred)
r2=r2_score(y6_test,y6_pred)
```

Performance Metrics

```python
print(f"Mean Squared Error (MSE):{mse}")
print(f"Root Mean Squared Error (RMSE):{rmse}")
print(f"Mean Absolute Error (MAE):{mae}")
print(f"R-squared (RÂ²):{r2}")
```
